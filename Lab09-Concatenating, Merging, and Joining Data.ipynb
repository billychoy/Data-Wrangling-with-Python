### Task 1 (3 marks)

Join the three dataframes using the _outer_ and _inner_ mechanisms via the __join__, __concat__, and __merge__ methods. The output from these three methods should look exactly the same.

#### Outer Join

```python
### Write your code below this comment.
df_left.index.names = ['key1']
df_right1.index.names = ['key1']
df_right2.index.names = ['key1']

df_outter = pd.merge(df_left, df_right1, how ="outer", on='key1')
print(df_outter)

df_outter2 = pd.merge(df_outter, df_right2, how ="outer", on='key1')
print('\n', df_outter2)
```

```python
#Concate - outter join
df_concate = pd.concat([df_left, df_right1, df_right2], axis=1, join="outer")
print(df_concate)
```

```python
#Simple join
df_join = df_left.join([df_right1, df_right2], how="outer")
print(df_join)
```

#### Inner Join

```python
### Write your code below this comment.
df_inner = pd.merge(df_left, df_right1, on='key1', how ="inner")
print(df_inner)

df_inner2 = pd.merge(df_inner, df_right2, how ="inner", on='key1')
print('\n', df_inner2)
```

```python
#Concate - inner join
df_concate = pd.concat([df_left, df_right1, df_right2], axis=1, join="inner")
print(df_concate)
```

```python
#Simple join
df_join = df_left.join([df_right1, df_right2], how="inner")
print(df_join)
```

```python
### Write your code below this comment.
def column_chk (tb1, tb2):
    results=[] 
    for x in tb1.columns:
        for y in tb2.columns:
            if any(tb1[x].isin(tb2[y])):
                results.append(tuple((x, y)))
    return results

print("The possible pairs are:")
print(column_chk(df_surveys, df_species))
```
